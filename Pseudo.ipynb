```python
Initialize parameters:
initial_wealth_A = 1000
initial_wealth_B = 2000
f = 0.1	# Fraction of wealth bet
T = 100	# Number of rounds
N = 10000	# Number of simulations

Initialize wealth arrays:
wealth_A = Array of size N filled with initial_wealth_A
wealth_B = Array of size N filled with initial_wealth_B

For each time step t from 0 to T-1 do:
Generate outcomes for Player A:
outcomes_A = Array of size N with each element randomly chosen to be either (1 + f) or (1 - f) with equal probability

Generate outcomes for Player B:
outcomes_B = Array of size N with each element randomly chosen to be either (1 + f) or (1 - f) with equal probability

Update wealth for Player A:
wealth_A = wealth_A multiplied element-wise by outcomes_A

Update wealth for Player B:
wealth_B = wealth_B multiplied element-wise by outcomes_B

Calculate time average growth rate for Player A (g_t_A):
g_t_A = Mean of log(wealth_A / initial_wealth_A)

Calculate time average growth rate for Player B (g_t_B):
g_t_B = Mean of log(wealth_B / initial_wealth_B)

Calculate ensemble average growth rate for Player A (g_e_A):
g_e_A = Log(mean of wealth_A / initial_wealth_A)

Calculate ensemble average growth rate for Player B (g_e_B):
g_e_B = Log(mean of wealth_B / initial_wealth_B)

Output the results
return g_t_A, g_e_A, g_t_B, g_e_B
```
